## 30.모델링1 : 단순회귀

* lstat(하위계층 비율)로만 집값(medv)을 예측하는 모델을 만들어 봅시다.

```
# 학습에 사용할 변수를 지정합니다.
features = ['lstat']

# 지정된 변수로 데이터셋을 만듭니다. 
# 학습과 검증시 아래 데이이터셋 이름을 사용합니다.
x_train1 = x_train[features]
x_val1 = x_val[features]
```

```
x_train1.head()
```

### 1)import

```
# 모델링용
from sklearn.linear_model import LinearRegression

# 회귀모델 평가용
from sklearn.metrics import mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

### 2) 모델선언

```
model1 = LinearRegression()
```

### 3) 모델링(학습)

```
model1.fit(x_train1, y_train)
```

* 회귀계수 확인하기

```
print('회귀계수 : ', model1.coef_)
print('절편 : ', model1.intercept_)
# medv = -0.9899*lstat + 35.065..
```

### 4) 검증 : 예측

```
x_val1[:5]
```

```
pred = model1.predict(x_val1) # lstat값 넣고 계산해서 나온 결과medv값 나오게 해라가 pred.에서 하는것 
pred
```

### 5) 검증 : 평가

```
# MSE
mean_squared_error(  y_val  , pred   )
```

```
# RMSE
mean_squared_error(    y_val  , pred   , squared=False)
```

```
# MAE
mean_absolute_error(     y_val  , pred   )
```

```
# MAPE
mean_absolute_percentage_error(     y_val  , pred    )
```

```
# 1-MAPE = 정확도
1 - mean_absolute_percentage_error(     y_val  , pred    )
```

### 6) 실습

* 집값을 예측하는데, 중요한 요인으로 판단되는 feature로 모델을 만들어 봅시다.
    * model2 : feature1
    * model3 : feature2
    * model4 : feature3

```
features = ['chas']
x_train2 = x_train[features]
x_val2 = x_val[features]
model2 = LinearRegression()
model2.fit(x_train2,y_train)
pred = model2.predict(x_val2)
mean_squared_error(    y_val  , pred   , squared=False) # 8.65
mean_absolute_error(     y_val  , pred   ) # 6.82
mean_absolute_percentage_error(     y_val  , pred    ) #0.47
```

```
features = ['rm']
x_train3 = x_train[features]
x_val3 = x_val[features]
model3 = LinearRegression()
model3.fit(x_train3,y_train)
pred = model3.predict(x_val3)
mean_squared_error(    y_val  , pred   , squared=False) # 7.50
mean_absolute_error(     y_val  , pred   ) # 5.22
mean_absolute_percentage_error(     y_val  , pred    ) #0.3399
```

```
features = ['rad']
x_train4 = x_train[features]
x_val4 = x_val[features]
model4 = LinearRegression()
model4.fit(x_train4,y_train)
pred = model4.predict(x_val4)
mean_squared_error(    y_val  , pred   , squared=False) # 7.78
mean_absolute_error(     y_val  , pred   ) # 6.13
mean_absolute_percentage_error(     y_val  , pred    ) #0.379
```
