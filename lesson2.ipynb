```python
import datetime
class Time_:
    def __init__(self):
        self.utc_date = datetime.datetime.utcnow()
        self.ukd = self.utc_date + datetime.timedelta(hours= 9)    #utc_korea_date
        self.uad = self.utc_date + datetime.timedelta(hours= -5)
        self.uad_summer = self.utc_date + datetime.timedelta(hours= -4)    #utc_america_date
        self.initial_date = self.uad + datetime.datetime(2019,1,1)
        self.timelist = {
            "pre_start" : 4,
            "pre_end" : 9.5,
            "main_start" : 9.5,
            "main_end" : 16,
            "post_start" : 16,
            "post_end" : 18
        }
        self.summer_start_month = '4'
        self.summer_end_month = '10'
        self.year = datetime.datetime.utc.year()
    #     """
    #     썸머타임일 때와 아닐 때 현재 장시간 설정
    #     썸머타임 기간: self.uad가 매년3월 마지막일요일 ~ 10월 마지막일요일 사이일 때 장시간 설정
    #     썸머타임 기간 아닐 때: self.uad가 매년3월 마지막일요일 ~ 10월 마지막일요일 사이가 아닐 때 장시간 설정
    #     """
    # start_summer, end_summer에 self 필요할지?
    def summer_period(self):    
        a =  datetime.datetime(self.year,self.summer_start_month,1)
        b = datetime.datetime(self.year,self.summer_end_month ,1)
        start_summer = a - datetime.timedelta(days = a.isoweekday())
        end_summer = b - datetime.timedelta(days = b.isoweekday())

        if start_summer < self.uad < end_summer : 
            self.pre_start_time_summer = self.uad_summer + datetime.timedelta(self.timelist['pre_start'])
            self.pre_end_time_summer = self.uad_summer + datetime.timedelta(self.timelist['pre_end'])
            self.main_start_time_summer = self.uad_summer + datetime.timedelta(self.timelist['main_start'])
            self.main_end_time_summer = self.uad_summer + datetime.timedelta(self.timelist['main_end'])       
            self.post_start_time_summer = self.uad_summer + datetime.timedelta(self.timelist['post_start'])
            self.post_end_time_summer = self.uad_summer + datetime.timedelta(self.timelist['post_end'])
        elif self.uad < start_summer or self.uad > end_summer :
            self.pre_start_time = self.uad + datetime.timedelta(self.timelist['pre_start'])
            self.pre_end_time = self.uad + datetime.timedelta(self.timelist['pre_end'])
            self.main_start_time = self.uad + datetime.timedelta(self.timelist['main_start'])
            self.main_end_time = self.uad + datetime.timedelta(self.timelist['main_end'])       
            self.post_start_time = self.uad + datetime.timedelta(self.timelist['post_start'])
            self.post_end_time = self.uad + datetime.timedelta(self.timelist['post_end'])
        return start_summer & end_summer
    

t = Time_()
print(t.summer_period())
# print(start_summer, end_summer)
     
    



```
